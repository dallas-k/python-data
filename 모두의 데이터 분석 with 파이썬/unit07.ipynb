```python
import csv
import matplotlib.pyplot as plt

target = input('읍면동 단위를 입력해주세요 : ')

f = open('population.csv')
data = csv.reader(f)
next(data)

result = []
for row in data :
    if target in row[0] :
        for i in row[3:] :
            result.append(int(i))

plt.plot(result)
plt.show()
```
