```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []

name = input('찾고 싶은 지역의 이름을 알려주세요 :')
for row in data :
    if name in row[0] :
        for i in row[3:14] :
            m.append(-int(i))
        for i in row[16:] :
            f.append(int(i))
        break
  
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize=(10,5), dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name + ' 남녀 성비')
plt.barh(range(len(m)), m, label='남성')
plt.barh(range(len(f)), f, label='여성')
plt.legend()
plt.show()
```

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
size = []
name = input('location? ')

for row in data :
    if name in row[0] :
        m = 0
        f = 0
        for x in row[3:14]:
            m += int(x)
        for y in row[16:] :
            f += int(y)
        break

size.append(m)
size.append(f)


import matplotlib.pyplot as plt
plt.pie(size, labels=['m','f'], autopct="%.1f%%", startangle=90)
plt.show()
```
