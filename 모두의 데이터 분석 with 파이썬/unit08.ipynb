```python
import csv
f = open('population.csv')
data = csv.reader(f)
next(data)

result = []

for row in data :
    if '신도림' in row[0] :
        for i in row[3:] :
            result.append(int(i))
            
import matplotlib.pyplot as plt
plt.bar(range(101), result)
plt.show()
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
next(data)

m = []
f = []

for row in data :
    if '신도림' in row[0] :
        for x in row[3:14] :
            m.append(-int(x))
        for y in row[16:] :
            f.append(int(y))

import matplotlib.pyplot as plt
plt.title('신도림 지역의 남녀 성비')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(11), m, label='male')
plt.barh(range(11), f, label='female')
plt.legend()
plt.show()
```
