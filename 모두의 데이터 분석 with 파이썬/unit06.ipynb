```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        result.append(float(row[3]))

plt.hist(result, bins=100, color='r')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data :
    month = row[2].split('-')
    if row[3] != '' :
        if month[1] == '01' :
            jan.append(float(row[3]))
        if month[1] == '08' :
            aug.append(float(row[3]))

plt.hist(aug, bins=30, color='r', label='August')
plt.hist(jan, bins=30, color='b', label='January')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    month = row[2].split('-')
    if row[3] != '' :
        result[int(month[1]) - 1].append(float(row[3]))
        
plt.boxplot(result)
plt.show()
```
