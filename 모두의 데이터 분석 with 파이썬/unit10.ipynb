```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = []

name = input('location : ')

for row in data :
    if name in row[0] :
        for x in range(11) :
            result.append(int(row[x+3]) - int(row[x+13]) )
        break
        
import matplotlib.pyplot as plt
plt.bar(range(11), result)
plt.show()
```
